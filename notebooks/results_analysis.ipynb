Celda 2: Visualización del Mosaico de Validación (Ground Truth vs. Predicción)

```python
# Definición de la ruta del mosaico visual de validación
val_mosaic_path = '../assets/val_batch0_pred.jpg' # Ajusta el nombre según lo tengas en assets

if os.path.exists(val_mosaic_path):
    img = mpimg.imread(val_mosaic_path)
    plt.figure(figsize=(14, 14))
    plt.imshow(img)
    plt.axis('off')
    plt.title("Mosaico de Validación: Inspección Visual de Instancias Segmentadas", fontsize=14, pad=15)
    plt.show()
else:
    print(f"Aviso: Coloca la imagen de resultados del lote de validación en: {val_mosaic_path}")
```

```python
Celda 3: Evaluación Analítica (Matriz de Confusión y Curva F1-Score)
```

```python
# Rutas a los gráficos analíticos exportados por YOLOv8
confusion_matrix_path = '../assets/confusion_matrix.png'
f1_curve_path = '../assets/F1_curve.png'

fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# 1. Despliegue de la Matriz de Confusión
if os.path.exists(confusion_matrix_path):
    img_cm = mpimg.imread(confusion_matrix_path)
    axes[0].imshow(img_cm)
    axes[0].axis('off')
    axes[0].set_title("Matriz de Confusión del Modelo", fontsize=12)
else:
    axes[0].text(0.5, 0.5, 'Matriz de confusión no encontrada\nen assets/', ha='center', va='center')
    axes[0].axis('off')

# 2. Despliegue de la Curva F1-Score
if os.path.exists(f1_curve_path):
    img_f1 = mpimg.imread(f1_curve_path)
    axes[1].imshow(img_f1)
    axes[1].axis('off')
    axes[1].set_title("Curva F1-Score vs. Confianza", fontsize=12)
else:
    axes[1].text(0.5, 0.5, 'Curva F1-Score no encontrada\nen assets/', ha='center', va='center')
    axes[1].axis('off')

plt.tight_layout()
plt.show()
```
