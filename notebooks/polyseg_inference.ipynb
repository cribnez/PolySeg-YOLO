Celda 1: Inicialización de Módulos y Entorno

```python
import os
import cv2
import time
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Desactivar avisos innecesarios en la consola
os.environ["YOLO_VERBOSE"] = "False"

print("=========================================================")
print("  ENTORNO DE INFERENCIA EN TIEMPO REAL: POLYSEG-YOLO     ")
print("=========================================================")
```

Celda 2: Carga del Modelo Optimizado

```python
# Ruta relativa hacia los pesos entrenados del proyecto
weights_path = '../weights/best.pt'

if os.path.exists(weights_path):
    # Carga de la arquitectura YOLOv8 con tus pesos optimizados
    model = YOLO(weights_path)
    print(f"Éxito: Cerebro del modelo cargado desde '{weights_path}'")
    print("PolySeg-YOLO se encuentra listo para la segmentación de instancias.")
else:
    print(f"Error crítico: No se encontró el archivo de pesos en {weights_path}")
    print("Asegúrate de descargar 'best.pt' desde Colab y colocarlo en la carpeta 'weights/'.")
```

Celda 3: Ejecución de Inferencia y Monitoreo de Latencia

```python
# Ruta hacia la imagen muestra para validar la inferencia
test_image_path = '../assets/sample_prediction.jpg'

if os.path.exists(test_image_path):
    print("Iniciando procesamiento de la muestra endoscópica...")
    
    # 1. Medición de tiempo estricta para evaluar la latencia (Benchmark local)
    start_time = time.time()
    
    # Ejecución de la predicción (Confianza mínima del 50% para evitar falsos positivos)
    # Cambiar device='cpu' por device=0 si cuentas con una GPU dedicada de NVIDIA
    results = model.predict(source=test_image_path, conf=0.50, device='cpu', verbose=False)
    
    # Cálculo final del tiempo transcurrido convertido a milisegundos (ms)
    execution_latency = (time.time() - start_time) * 1000
    
    print("\n---------------------------------------------------------")
    print(f" VALOR DE LATENCIA EVALUADO: {execution_latency:.2f} ms")
    print("---------------------------------------------------------")
    
    # 2. Despliegue visual de los mapas de segmentación y contornos geométricos
    for result in results:
        # El método .plot() dibuja automáticamente contornos, máscaras e ID de clase
        annotated_frame = result.plot()
        
        # Ajustar dimensiones de visualización en el cuaderno
        plt.figure(figsize=(10, 10))
        
        # OpenCV lee en formato BGR; convertimos a RGB para que Matplotlib muestre los colores reales
        plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title(f"Segmentación Computarizada (Latencia: {execution_latency:.2f} ms)", fontsize=12, pad=10)
        plt.show()
else:
    print(f"Aviso: Para ejecutar la prueba de inferencia, añade una imagen de pólipo en: {test_image_path}")
```
